##Data Visualization and Analysis

*We are changing and generalizing the date format*

```python
df['Date']=pd.to_datetime(df['Date'])
df
```

*We are analysing and classifying the data on the basis of whether they are in Pre-Covid or Covid Period*

```python
covid_start_date='17-03-2020'
pre_covid_df=df[df['Date']<covid_start_date]
during_covid_df= df[df['Date']>=covid_start_date]
print('Pre-Covid Period')
print(pre_covid_df)
print('During Covid Period')
print(during_covid_df)
```

*Summary of the datasets in Pre-COVID and during COVID period.*

```python
print("Summary of Pre-COVID Period:")
print(pre_covid_df.describe())
print("\nSummary During COVID Period:")
print(during_covid_df.describe())
```

*The summary of Unemployment Rate in India in the Pre-COVID period and during COVID period can be visualized using the following bar graph.*

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=pre_covid_df, color='blue', alpha=0.5, label='Pre-COVID')
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=during_covid_df, color='red', alpha=0.5, label='During COVID')
plt.xticks(rotation=90)
plt.title('Unemployment Rate by Region: Pre-COVID vs During COVID')
plt.legend()
plt.show()
```

####**Some Important Observations from the bar graph**

The x-axis represents the Regions while the y axis the Estimated Unemployment Rate.

In most regions, the unemployment rate during COVID-19 (red bars) is higher compared to the pre-COVID period (blue bars). This indicates a general increase in unemployment rates due to the pandemic.

Certain regions like Bihar, Jharkhand, Puducherry and Tamil Nadu show a substantial increase in unemployment rates during the COVID-19 period compared to others.

Regions like Chattisgarh, Gujarat, Meghalaya, Sikkim and Uttar Pradesh show a moderate increase in unemployment rates, indicating that while there was an impact, it was not as severe as in other regions.

Regions like Assam, Himachal Pradesh, Chandigarh, Jammu and Kashmir and Tripura's unemployment rates were unaffected during Covid-19.

The maximum unemployment rate observed during COVID-19 (76.74%) is almost double the pre-COVID maximum (34.69%).
